## A simple agent using tools from the MCP server


```python
import os
from langchain_core.tools import tool
from langchain_core.messages import HumanMessage
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_mcp_adapters.client import MultiServerMCPClient

# ---------------------------
# 1. Define a local tool
# ---------------------------
@tool
def say_hello(name: str) -> str:
    """Say hello to a person by name."""
    return f"Hello, {name}! ğŸ‘‹"

# ---------------------------
# 2. Load MCP tools + merge
# ---------------------------
client = MultiServerMCPClient({
    "social_graph": {
        "transport": "http",
        "url": "https://ftec5660.ngrok.app/mcp",
        "headers": {"ngrok-skip-browser-warning": "true"}
    }
})

mcp_tools = await client.get_tools()
tools = mcp_tools + [say_hello]

# ---------------------------
# 3. Initialize Gemini (tool-enabled) or deepseek
# ---------------------------
llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    google_api_key=GEMINI_VERTEX_API_KEY,
    temperature=0,
)

# from langchain_openai import ChatOpenAI
# DEEPSEEK_API_KEY = userdata.get("DEEPSEEK_API_KEY")
# llm = ChatOpenAI(
#     model="deepseek-chat",          # or "deepseek-reasoner"
#     api_key=DEEPSEEK_API_KEY,
#     base_url="https://api.deepseek.com/v1",
#     temperature=0,
# )

llm_with_tools = llm.bind_tools(tools)

# ---------------------------
# 4. Single-step invocation
# ---------------------------
from langchain_core.messages import HumanMessage, ToolMessage

query = "Say hello to Bao using tool, then search for someone named Alice on Facebook."

messages = [HumanMessage(content=query)]
ai_msg = llm_with_tools.invoke(messages)

print("First AI message:", ai_msg)
print("Tool calls:", getattr(ai_msg, "tool_calls", None))

if getattr(ai_msg, "tool_calls", None):
    messages.append(ai_msg)

    tool_map = {t.name: t for t in tools}

    for tc in ai_msg.tool_calls:
        tool_name = tc["name"]
        tool_args = tc.get("args", {})
        tool_id = tc["id"]

        print(f"Running tool: {tool_name} with args={tool_args}")

        tool_obj = tool_map[tool_name]

        try:
            result = await tool_obj.ainvoke(tool_args)
        except Exception:
            result = tool_obj.invoke(tool_args)

        print(f"Tool result ({tool_name}):", result)

        messages.append(
            ToolMessage(
                content=str(result),
                tool_call_id=tool_id
            )
        )

    final_msg = llm_with_tools.invoke(messages)
    print("\nFinal answer:")
    print(final_msg)
else:
    print("No tool calls were generated.")
```

# Evaluation code

In the test phase, you will be given 5 CV files with fixed names:

    CV_1.pdf, CV_2.pdf, CV_3.pdf, CV_4.pdf, CV_5.pdf

Your system must process these CVs and output a list of 5 scores,
one score per CV, in the same order:

    scores = [s1, s2, s3, s4, s5]

Each score must be a float in the range [0, 1], representing the
reliability or confidence that the CV is valid (or meets the task criteria).

The ground-truth labels are binary:

    groundtruth = [0 or 1, ..., 0 or 1]

Each CV is evaluated independently using a threshold of 0.5:

- If score > 0.5 and groundtruth == 1 â†’ Full credit
- If score â‰¤ 0.5 and groundtruth == 0 â†’ Full credit
- Otherwise â†’ No credit

In other words, 0.5 is the decision threshold.

- Each CV contributes equally.
- Final score = (number of correct decisions) / 5


```python
# =====================================================
#  Evaluation code
# =====================================================
from langchain_core.messages import HumanMessage

def norm(s):
    return re.sub(r"\s+", " ", str(s).lower()).strip()

def to_text(res):
    if isinstance(res, str):
        return res
    if isinstance(res, list):
        out = []
        for x in res:
            if isinstance(x, dict) and x.get("type") == "text":
                out.append(str(x.get("text", "")))
            else:
                out.append(str(x))
        return "\n".join(out)
    return str(res)

def _json_from_llm(prompt, default):
    try:
        r = llm.invoke([HumanMessage(content=prompt)])
        txt = r.content if hasattr(r, "content") else str(r)
        if isinstance(txt, list):
            txt = "\n".join(str(x) for x in txt)
        txt = str(txt).strip()
        try:
            return json.loads(txt)
        except:
            m = re.search(r"\{.*\}|\[.*\]", txt, re.S)
            return json.loads(m.group(0)) if m else default
    except:
        return default

def extract_claims(cv_text):
    prompt = f"""
Extract CV info. Return ONLY JSON:
{{
  "name": string|null,
  "roles": [string],
  "locations": [string]
}}
Rules:
- roles should be high-level (AI/Finance/Design/Education/Logistics/Marketing/etc.)
- no guessing
CV:
{cv_text[:8000]}
"""
    d = _json_from_llm(prompt, {"name": None, "roles": [], "locations": []})
    # fallback name
    if not d.get("name"):
        for ln in [x.strip() for x in cv_text.splitlines() if x.strip()][:12]:
            t = " ".join(re.sub(r"\|", " ", ln).split())
            if re.fullmatch(r"[A-Za-z][A-Za-z\-\.\s]{2,60}", t) and 1 <= len(t.split()) <= 4:
                if norm(t) not in {"professional experience","education","skills","projects","summary"}:
                    d["name"] = t
                    break
    d["roles"] = d.get("roles", []) or []
    d["locations"] = d.get("locations", []) or []
    return d

def name_score(a, b):
    a, b = norm(a), norm(b)
    if not a or not b: return 0.0
    if a == b: return 1.0
    if a in b or b in a: return 0.8
    sa, sb = set(a.split()), set(b.split())
    return len(sa & sb) / max(1, len(sa))

async def best_linkedin(claims):
    if not claims.get("name"): return None, None
    try:
        raw = await search_linkedin_people.ainvoke({"q": claims["name"], "limit": 5, "fuzzy": True})
        cands = parse_mcp_list_result(raw)
    except Exception as e:
        print("[LI search]", e); return None, None
    if not cands: return None, None
    best = sorted(cands, key=lambda c: (
        2*name_score(claims["name"], c.get("name","")) +
        (0.2 if str(c.get("match_type","")).lower()=="exact" else 0)
    ), reverse=True)[0]
    prof = None
    try:
        pid = best.get("id")
        if isinstance(pid, str) and pid.isdigit(): pid = int(pid)
        if isinstance(pid, int):
            prof = await get_linkedin_profile.ainvoke({"person_id": pid})
    except Exception as e:
        print("[LI profile]", e)
    return best, prof

async def best_facebook(claims):
    if not claims.get("name"): return None, None
    try:
        raw = await search_facebook_users.ainvoke({"q": claims["name"], "limit": 5, "fuzzy": True})
        cands = parse_mcp_list_result(raw)
    except Exception as e:
        print("[FB search]", e); return None, None
    if not cands: return None, None
    best = sorted(cands, key=lambda c: (
        2*name_score(claims["name"], c.get("display_name","")) +
        (0.2 if str(c.get("match_type","")).lower()=="exact" else 0)
    ), reverse=True)[0]
    prof = None
    try:
        uid = best.get("id")
        if isinstance(uid, str) and uid.isdigit(): uid = int(uid)
        if isinstance(uid, int):
            prof = await get_facebook_profile.ainvoke({"user_id": uid})
    except Exception as e:
        print("[FB profile]", e)
    return best, prof

def llm_score(claims, li_best, li_prof, fb_best, fb_prof):
    prompt = f"""
Judge whether the CV is valid based on social profiles. Return ONLY JSON:
{{"score": 0.0-1.0, "reason": "short"}}
Use 0.5 as threshold. Be conservative.
Use LinkedIn as PRIMARY evidence for professional claims (role, industry, years_experience).
Use Facebook as SECONDARY evidence for identity/location/social consistency only.
Do NOT heavily penalize a CV if LinkedIn strongly supports the CV but Facebook conflicts.
Penalize strongly only when BOTH LinkedIn and Facebook contradict core CV claims, or when LinkedIn strongly contradicts core professional claims.
If name matches exactly on both LinkedIn and Facebook, and at least one CV location is supported by either profile,
the score should usually be above 0.5 unless LinkedIn strongly contradicts the core profession and both profiles also contradict location.

Scoring guidance:
- 0.75-0.95: LinkedIn strongly supports core claims; Facebook may be partial/missing
- 0.55-0.75: partial support with some inconsistencies
- 0.25-0.50: major inconsistencies / weak support
- 0.00-0.25: likely invalid / severe contradictions

CV claims:
{json.dumps(claims, ensure_ascii=False)}

LinkedIn candidate:
{json.dumps(li_best or {}, ensure_ascii=False)}

Facebook candidate:
{json.dumps(fb_best or {}, ensure_ascii=False)}

LinkedIn profile text:
{to_text(li_prof)[:1800]}

Facebook profile text:
{to_text(fb_prof)[:1200]}
"""
    d = _json_from_llm(prompt, {"score": 0.5, "reason": "fallback"})
    try:
        s = float(d.get("score", 0.5))
    except:
        s = 0.5
    # compress to avoid too many 1.0
    s = 0.05 + 0.90 * max(0.0, min(1.0, s))
    return round(s, 4), d.get("reason", "")

def cv_sort_key(item):
    m = re.search(r"(\d+)", item["file"])
    return int(m.group(1)) if m else 999

# ---- Run scoring on CV_1..CV_5 ----
scores = []
for cv_item in sorted(all_cvs, key=cv_sort_key)[:5]:
    claims = extract_claims(cv_item["text"])
    li_best, li_prof = await best_linkedin(claims)
    fb_best, fb_prof = await best_facebook(claims)
    s, reason = llm_score(claims, li_best, li_prof, fb_best, fb_prof)
    scores.append(s)

    print("="*60)
    print(cv_item["file"], "score=", s)
    print("claims:", claims)
    print("LI:", li_best)
    print("FB:", fb_best)
    print("reason:", reason)

print("scores =", scores)

def evaluate(scores, groundtruth, threshold=0.5):
    """
    scores: list of floats in [0, 1], length = 5
    groundtruth: list of ints (0 or 1), length = 5
    """
    assert len(scores) == 5
    assert len(groundtruth) == 5

    correct = 0
    decisions = []

    for s, gt in zip(scores, groundtruth):
        pred = 1 if s > threshold else 0
        decisions.append(pred)
        if pred == gt:
            correct += 1

    final_score = correct / len(scores)

    return {
        "decisions": decisions,
        "correct": correct,
        "total": len(scores),
        "final_score": final_score
    }
```
